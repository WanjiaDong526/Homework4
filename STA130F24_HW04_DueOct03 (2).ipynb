````python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (you can replace this with your dataset)
data = np.array([5.2, 5.5, 5.7, 4.9, 5.3, 5.8, 5.1, 5.4])

# Parameters
n_iterations = 10000  # Number of bootstrap samples
n = len(data)         # Number of observations in the original dataset

# List to store the medians of bootstrap samples
medians = []

# Bootstrapping process
for _ in range(n_iterations):
    # Generate a bootstrap sample with replacement
    sample = np.random.choice(data, size=n, replace=True)
    # Calculate the median of the bootstrap sample and append it to the list
    medians.append(np.median(sample))

# Convert medians list to a numpy array for easier calculations
bootstrapped_medians = np.array(medians)

# Calculate the 95% confidence interval
lower_bound = np.percentile(bootstrapped_medians, 2.5)
upper_bound = np.percentile(bootstrapped_medians, 97.5)

# Output the results
print(f"Bootstrapped Median: {np.median(bootstrapped_medians)}")
print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")

# Optional: Plot the distribution of bootstrapped medians
plt.hist(bootstrapped_medians, bins=30, alpha=0.7, color='blue')
plt.axvline(lower_bound, color='red', linestyle='dashed', linewidth=2, label='Lower 95% CI')
plt.axvline(upper_bound, color='green', linestyle='dashed', linewidth=2, label='Upper 95% CI')
plt.axvline(np.median(data), color='yellow', linestyle='dashed', linewidth=2, label='Sample Median')
plt.title('Bootstrapped Medians Distribution')
plt.xlabel('Median Values')
plt.ylabel('Frequency')
plt.legend()
plt.savefig("my_plot.png")
plt.show()
````

Population parameters are numerical values that describe the characteristics of the population and sample statistic are numerical values that are calculated from samples drawn from the population.
The population parameters are the true values we wish to understand, and the sample statistics are the best estimates we can make based on the collected data. This distinction emphasizes the uncertainty in estimating population features from finite samples.

Summaries:

Difference Between Population Parameter and Sample Statistic:

Population Parameter: A fixed value summarizing a characteristic of the entire population, often unknown (e.g., mean, standard deviation). Sample Statistic: A value calculated from a sample that estimates the population parameter, variable based on sample data. Importance in Confidence Intervals: Sample statistics are used to create confidence intervals, which estimate the range within which the population parameter likely falls, quantifying uncertainty in estimates.
